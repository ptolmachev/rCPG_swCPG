```python
import numpy as np
from collections import deque
from copy import deepcopy
from matplotlib import pyplot as plt
```

```python
class HCO():
    def __init__(self, dt, drives, weights):
        self.weights = weights
        self.drives = drives
        self.dt = dt
        self.vars = ["v1", "v2", "m1", "m2"]
        self.state = np.random.rand(4)
        self.tau = 1500
        self.state_history = deque(maxlen = 200000)
    
    def fr(self, v):
        return 1.0 / (1 + np.exp(-v))
    
    def rhs(self):
        v1 = self.state[self.vars.index('v1')]
        v2 = self.state[self.vars.index('v2')]
        m1 = self.state[self.vars.index('m1')]
        m2 = self.state[self.vars.index('m2')]
        w12, w21 = self.weights
        
        rhs_v1 = -0.01 * v1 - m1 - w21 * self.fr(v2) + self.drives[0]
        rhs_v2 = -0.01 * v2 - m2 - w12 * self.fr(v1) + self.drives[1]
        rhs_m1 = (self.fr(v1) - m1) / self.tau
        rhs_m2 = (self.fr(v2) - m2) / self.tau
        return np.array([rhs_v1, rhs_v2, rhs_m1, rhs_m2])
    
    def step(self):
        self.state += self.dt * self.rhs()
        self.state_history.append(deepcopy(self.state))
        return None
        
    def run(self, T_steps):
        for i in range(T_steps):
            self.step()
        return None
```

```python
dt = 0.1
drives = 0.5 * np.ones(2)
weights = 0.25 * np.ones(2)
hco = HCO(dt=dt, drives=drives, weights=weights)
hco.run(80000)
```

```python
state_history = np.array(hco.state_history)
fig = plt.figure(figsize = (14, 7))
plt.plot(state_history[:, 0], label = "First Neuron")
plt.plot(state_history[:, 1], label = "Second Neuron")
plt.legend(fontsize = 15)
plt.title("Half-centre oscillator dynamics", fontsize=25)
plt.xlabel("t", fontsize=15)
plt.ylabel("v", fontsize=15)
plt.show()
```

```python
class HCO():
    def __init__(self, dt, drives, weights, tau):
        self.weights = weights
        self.drives = drives
        self.dt = dt
        self.tau = tau
        self.state_history = deque(maxlen = 100000)
        self.nrn1 = Neuron(dt=self.dt, drive=drives[0], tau=tau)
        self.nrn2 = Neuron(dt=self.dt, drive=drives[1], tau=tau)
    
    def fr(self, v):
        return 1.0 / (1 + np.exp(-v))
    
    def get_next_state(self):
        vs = np.array([self.nrn1.state[0], self.nrn2.state[0]])
        # calculate synaptic inhibition between the two neurons
        synaptic_inhibition = -(self.weights * self.fr(vs))[::-1]
        state1 = self.nrn1.state + self.dt * self.nrn1.rhs(synaptic_inhibition[0])
        state2 = self.nrn2.state + self.dt * self.nrn2.rhs(synaptic_inhibition[1])
        return np.hstack([state1, state2])
    
    def step(self):
        # update the state of hco
        self.state = self.get_next_state()
        # update the state of infividual neurons
        self.nrn1.state = deepcopy(self.state[:2])
        self.nrn2.state = deepcopy(self.state[2:])
        # save values to the history
        self.nrn1.state_history.append(deepcopy(self.nrn1.state))
        self.nrn2.state_history.append(deepcopy(self.nrn2.state))
        return None
    
    def get_state_history(self):
        return np.hstack([self.nrn1.get_state_history(), self.nrn2.get_state_history()])
    
    def run(self, T_steps):
        for i in range(T_steps):
            self.step()
        return None
```

```python
dt = 0.1
drives = 0.5 * np.ones(2)
weights = 0.25 * np.ones(2)
tau = 1500
hco = HCO(dt=dt, drives=drives, weights=weights, tau=tau)
```

```python
hco.run(100000)
```
